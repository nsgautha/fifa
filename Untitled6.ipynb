```python
country = fifa_data["Nationality"].value_counts() 
main_country = country.idxmax()
main_country_count = country.max()
```

```python
print(f"The country with the most number of players is {main_country} with {main_country_count} players.")
```

```python
#plotting bar graph
top_5_countries = country.head(5)
plt.figure(figsize=(8, 4))
top_5_countries.plot(kind='bar')
plt.title('Top 5 Countries with Most Number of Players')
plt.xlabel('Country')
plt.ylabel('No of Players')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
fifa_data = pd.read_csv("fifa_data.csv")
fifa_data
```

```python
#Which player has the highest salary?
fifa_data['Wage'] = fifa_data['Wage'].astype(str).str.replace('€','').str.replace('K','000').astype(float)
high_salary_player = fifa_data.loc[fifa_data['Wage'].idxmax(), ['Name', 'Wage']]
high_salary_player_name = high_salary_player['Name']
high_salary_player_wage = high_salary_player['Wage']
print("The player with the highest salary is", high_salary_player)
```

```python
# 4. Plot a histogram to get the salary range of the players.
plt.figure(figsize=(8,4))
plt.hist(fifa_data['Wage'],color='g')
plt.title('Salary Range of Players')
plt.xlabel('Wage (€)')
plt.ylabel('Number of Players')
plt.show()
```

```python
# 7. Which foot is most preferred by the players? Draw a bar chart for preferred foot.
preferred_foot = fifa_data['Preferred Foot'].value_counts()
plt.figure(figsize=(8, 4))
preferred_foot.plot(kind='bar')
plt.title('Preferred Foot of Players')
plt.xlabel('Foot')
plt.ylabel('Number of Players')
plt.show()
```
